## What Heretic Does Automatically

For hybrid SSM/Transformer models like Falcon H1R:

1. **Detects Components**: Automatically finds both Transformer attention (`attn.o_proj`) and SSM/Mamba (`ssm.out_proj`) components
2. **Computes Refusal Directions**: Calculates directions in the residual space that represent censorship behavior
3. **Optimizes Parameters**: Uses Bayesian optimization to find the best ablation weights for each component type
4. **Applies Abliteration**: Orthogonalizes the weight matrices with respect to refusal directions using LoRA adapters

## Memory Requirements

- **Falcon H1R 7B**: ~16GB RAM for CPU inference
- **GPU**: 8GB+ VRAM recommended for faster processing
- **Optimization**: Heretic automatically adjusts batch sizes and uses hybrid CPU/GPU memory when available
